```python
import requests
import json
import pandas as pd
from pymongo import MongoClient
import mplfinance as mpf
```

```python
client = MongoClient('mongodb://localhost:27017/')
```

```python
stock_db = client['Project_3_Group_4']
```

```python
mongo_collection = stock_db['Stock_data']
```

DWTX Query

```python
dwtx_query = {'symbol':"DWTX"}
dwtx = list(mongo_collection.find(dwtx_query))
for d in dwtx:
    print(d)

dwtx_df = pd.DataFrame(dwtx)
#dwtx_df.to_csv('Stock_data/dwtx_data.json', index=False)
```

```python
dwtx_df
```

```python
dwtx_df = dwtx_df.drop(columns=["_id"])
dwtx_df = dwtx_df.rename(columns={"1. open": "Open", "2. high": "High", "3. low": "Low", "4. close": "Close", "5. volume": "Volume", "symbol":"Symbol", "date":"Date"})

dwtx_df
dwtx_df.dtypes
```

```python
#dwtx_df['date'] = pd.to_datetime(dwtx_df['date'])
dwtx_df.dtypes
```

```python
dwtx_df['Open'] = dwtx_df['Open'].astype(float)
dwtx_df['High'] = dwtx_df['High'].astype(float)
dwtx_df['Low'] = dwtx_df['Low'].astype(float)
dwtx_df['Close'] = dwtx_df['Close'].astype(float)
dwtx_df['Volume'] = dwtx_df['Volume'].astype(float)
```

```python
dwtx_df.dtypes
```

```python
dwtx_df = dwtx_df.sort_values(by="Date", ascending=True)
```

```python
dwtx_df.set_index('Date', inplace=True)
dwtx_df.shape
dwtx_df.head(5)
```

```python
dwtx_df = dwtx_df.reset_index(drop=False)
dwtx_df
```

```python
dwtx_df.to_json('Stock_data/dwtx_data.json', orient='records',index=4)
```
